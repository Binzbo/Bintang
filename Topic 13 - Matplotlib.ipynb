# Matplotlib untuk Numpy Array

Salah satu struktur data yang didukung oleh matplotlib adalah numpy array (ndarray). Untuk bisa memvisualisasikan diagram dari numpy array, gunakan perintah berikut:

```python

```

Mari kita coba kode program berikut:

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
x = np.linspace(0, 5, 11) # spaced number di numpy array, baca topik 9
y = x ** 2
````

````python
print(x)
````

````python
print(y)
````

````python
plt.plot(x, y, 'r') # 'r' adalah nilai warna merah
plt.xlabel('Judul sumbu X di sini')
plt.ylabel('Judul sumbu y di sini')
plt.title('Judul Diagram di sini')
plt.show()
````

# Method dari Matplotlib

Matplotlib merupakan sebuah class yang di dalamnya terdapat banyak method untuk mendukung proses visualisasi. Berikut adalah beberapa method yang sering digunakan pada matplotlib:

| Method | Kegunaan |
| :--- | :--- |
| figure() | Membuat canvas kosong |
| add_axes() | Menambahkan axes (satu set diagram sumbu) pada canvas |
| subplots() | Automatic axis manager (mirip seperti figure() tapi memiliki automasi dalam membuat axes) |
| tight_layout() | Men-adjust tampilan axes agar tidak saling tumpang tindih (overlapping) |
| savefig() | Men-generate output berkualitas tinggi dalam format PNG, JPG, EPS, SVG, PGF and PDF |
| set_title() | Memberikan judul pada diagram yang ditampilkan |
| legend() | Memunculkan legenda pada diagram yang ditampilkan |
| set_xlabel() | Memberikan label pada axis x |
| set_ylabel() | Memberikan label pada axis y |

## Penggunaan Methof Figure dan add_axes

Berikut adalah contoh dari penggunaan method ``figure()`` dan ``add_axes()``:

````python
fig = plt.figure()

# Menambah axes ke dalam canvas
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

# Plotting array x dan y ke dalam axes
axes.plot(x, y, 'b')
axes.set_xlabel('Set X Label') # Notice the use of set_ to begin methods
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
````

Kita juga bisa memunculkan multiaxes seperti pada kode program berikut:

````python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes

# Larger Figure Axes 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes2')
axes1.set_ylabel('Y_label_axes2')
axes1.set_title('Axes 2 Title')

# Insert Figure Axes 2
axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title');
````

## Penggunaan subplots

Berikut contoh penggunaan ``subplots()``:

````python
# Use similar to plt.figure() except use tuple unpacking to grab fig and axes
fig, axes = plt.subplots()

# Now use the axes object to add stuff to plot
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
````

Jika menggunakan ``subplots()`` maka axes akan di-generate otomatis. Kita juga bisa menentukan jumlah kolom dan baris jika ingin menampilkan multiplot. Mari kita coba kode program berikut:

````python
fig, axes = plt.subplots(nrows=1, ncols=2)
````

Penjelasan:

1. Kode program tersebut memunculkan dua axes dalam dua kolom dan satu baris sesuai dengan isian paramater. 
2. Hasil dari ``subplots()`` adalah tuple. Oleh karena itu kita menggunakan dua variabel untuk menampung return value dari fungsi tersebut.

Variabel ``axes`` digunakan untuk menampung array axes yang dibentuk dari ``subplots``. Berikut adalah isi dari axes:

````python
axes
````

Dikarenakan axes adalah sebuah numpy array, maka kita bisa melakukan iterasi menggunakan iterator dalam melakukan plotting pada masing-masing axes. Berikut contoh kode programnya:

````python
for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

# Menampilkan objek figure
fig
````

## Penggunaan tight_layout

Masalah yang umum terjadi di matplotlib ketika multiplot adalah adanya objek yang tumpang tindih di dalam canvas atau disebut juga sebagai <b>Overlapping</b>. Untuk menghindari hal tersebut kita bisa menggunakan fungsi ``tight_layout()`` pada figure atau plot. Berikut contoh penggunaanya:

````python
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

fig
plt.tight_layout()
````

# Ukuran Figure, Aspect Ratio, dan DPI

Matplotlib menyediakan fasilitas untuk kita menentukan aspect ratio, DPIm dan ukuran figure ketika object figure dibuat (menggunakan method ``figure()``). 

Penjelasan:

1. ``figsize`` berbentuk tuple yang berisikan width dan height dari sebuah figure dalam satuan inchi.
2. ``dpi`` singkatan dari dots-per-inch (jumlah pixel dalam setiap inchi). Makin tinggi dpi makin bagus kualitas tampilan grafiknya.

Berikut cara penggunaan parameter-parameter tersebut dalam method ``figure()``:

````python
fig = plt.figure(figsize=(8,4), dpi=100)
````

Parameter tersebut juga bisa digunakan pada fungsi ``subplots()``. Berikut contoh penggunaannya:

````python
fig, axes = plt.subplots(figsize=(12,3))

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
````

# Label, dan Titles

Setiap axes mempunyai beberapa komponen yang bisa dikustom, yaitu label, dan titles. Kedua komponen ini bisa digunakan untuk meningkatkan readibility dari visualisasi data yang ditampilkan. Berikut contoh penggunaan kedua komponen tersebut:

````python
tahun = ['2017', '2018', '2019', '2020', '2021']
jumlah_komoditi = [100, 200, 300, 300, 500]

plt.plot(tahun, jumlah_komoditi, 'b') # 'r' adalah nilai warna merah
plt.xlabel('Tahun')
plt.ylabel('Jumlah Komoditi')
plt.title('Jumlah Komoditi Beras di Indonesia dalam satuan Juta Ton')
plt.show()
````

# Legend

Untuk bisa menggunakan legend pada suatu visualisasi, kita bisa menggunakan parameter ``label = '<label_text>'`` pada saat melakukan plot atau di objek yang ditambahkan ke figure. Setelah itu panggil method ``legend()`` untuk memunculkan legend dari visualisasi tersebut. Berikut contoh penggunaannya:


````python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend()
````

Apabila legend overlapping dengan objek lainnya maka kita bisa memberikan parameter ``loc = <0..4>`` untuk menentukan posisi kemunculan legend pada figure. Penjelasan nilai ``loc`` dapat dilihat pada tabel berikut:

| Nilai loc | Keterangan |
| :--- | :--- |
| 1 | Pojok kanan atas |
| 2 | Pojok kiri atas |
| 3 | Pojok kiri bawah |
| 4 | Pojok kanan bawah |
| 0 | Matplotlib akan memilihkan posisi optimal untuk menempatkan label |


# Mengatur Warna, Linewidth, dan Jenis Garis

Kita bisa memilih warna, linewidth, dan jenis garis yang digunakan untuk mem-plotting data. Proses penentuan ini menggunakan konsep yang sama dengan konsep pada Matlab. Ketiga nilai untuk menentukan komponen tersebut diletakkan sebagai paramater pada fungsi ``plot()``. Urutannya adalah:

```python
'<nilai_warna><linewidth><jenis_garis>
```

Berikut contoh penggunaannya:

````python
fig, axes = plt.subplots()
axes.plot(x, x**2, 'b.-') # blue line with dots
axes.plot(x, x**3, 'g--') # green dashed line
````

Selain menggunakan konsep matlab dalam menentukan tiga komponen tersebut, kita bisa menggunakan mekanisme lain yaitu dengan cara memanggil keyword argument yang sesuai dengan paramter yang mau diubah. Berikut pembahasannya.

## Parameter color

Kita bisa menggunakan parameter color untuk menentukan warna dari suatu komponen visualisasi. Selain dengan menyebutkan nama dari warna yang mau digunakan, kita juga bisa menyebutkan warna menggunakan nilai hexa dari warna tersebut. Berikut adalah contoh penggunaannya:

````python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.5) # half-transparant
ax.plot(x, x+2, color="#8B008B")        # RGB hex code
ax.plot(x, x+3, color="#FF8C00")        # RGB hex code 
````

## Parameter linewidth dan linestyle

Untuk mengubah linewidth dan linestyle kita bisa menambahkan keyword argument berupa:

| Parameter | Keterangan |
| :--- | :--- |
| ``linewidth = `` atau ``lw = `` | linewidth |
| ``linestyle = `` atau ``ls = `` |  linestyle (nilainya berkisar antara ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’) |
| ``marker = `` | Memberikan market pada point data (nilainya berkisar antara ``'+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...``) |
| ``markersize = `` | Menentukan besar dari marker |
| ``markerfacecolor = `` | Menentukan warna dari marker |
| ``markeredgewidth = `` | Menentukan lebar dari border marker |
| ``markeredgecolor = `` | Menentukan warana border dari marker |

Berikut contoh penggunaannya:
